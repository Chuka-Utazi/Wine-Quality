```python
sns.scatterplot (x ='density', y='quality', data=df)
plt.title("WINE QUALITY AND DENSITY")
plt.show();
```

```python
ax = sns.countplot(x = df["quality"], order = df["quality"].value_counts(ascending = False).index)
values = df["quality"].value_counts(ascending = False).values
ax.bar_label(container=ax.containers[0], labels = values);
```

The Wines with a quality of 5 and 6 should be focused on more than the others. The company should save resources and focus only on the 5 and 6

```python
top_10_alcohol_counts = df["alcohol"].value_counts(ascending=False).head(10)

ax = sns.countplot(
    x="alcohol",
    order=top_10_alcohol_counts.index, 
    data=df  
)

ax.set_title("Count of Top 10 Alcohol Values")
ax.set_xlabel("Alcohol")
ax.set_ylabel("Count")

values = top_10_alcohol_counts.values
ax.bar_label(container=ax.containers[0], labels=values)

plt.show()
```

```python
top_10_alcohol_counts = df["alcohol"].value_counts(ascending=False).head(-10)

ax = sns.countplot(
    x="alcohol",
    order=top_10_alcohol_counts.index, 
    data=df  
)

ax.set_title("Count of Last 10 Alcohol Values")
ax.set_xlabel("Alcohol")
ax.set_ylabel("Count")

values = top_10_alcohol_counts.values
ax.bar_label(container=ax.containers[0], labels=values)

plt.show()
```

The Wine Company should only focus on the top 10 wines with a higher alcoholic percentage to save resources from the budget

```python
top_10_alcohol_counts = df["density"].value_counts(ascending=False).head(10)

ax = sns.countplot(
    x="density",
    order=top_10_alcohol_counts.index,  # Use the top 10 values as the order
    data=df  # Ensure you specify the DataFrame
)

ax.set_title("Count of Top 10 Density Values")
ax.set_xlabel("Density")
ax.set_ylabel("Count")

values = top_10_alcohol_counts.values
ax.bar_label(container=ax.containers[0], labels=values)

plt.show()
```
